# Travelling Salesman Problem using Dynamic Programming.
Problem Description:
Given a set of cities and distances, find the shortest possible route that visits each city once and returns to the starting city.

#Theory
TSP uses dynamic programming with bit masking. Store results of visited subsets to reduce repeated work.

## Algorithm
1. Start from source city
2. Mark visited cities using bit mask
3. Try all unvisited cities
4. Choose minimum cost path
5. Return to source city

```python
def tsp(graph):
    n = len(graph)
    dp = [[-1]*(1<<n) for _ in range(n)]

    def visit(city, mask):
        if mask == (1<<n) - 1:
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')

        for next_city in range(n):
            if mask & (1<<next_city) == 0:
                new_cost = graph[city][next_city] + visit(next_city, mask | (1<<next_city))
                if new_cost < ans:
                    ans = new_cost

        dp[city][mask] = ans
        return ans

    return visit(0, 1)
```

```python
graph = [
[0, 10, 15, 20],
[10, 0, 35, 25],
[15, 35, 0, 30],
[20, 25, 30, 0]
]

print("Minimum travelling cost:", tsp(graph))
```

```python
def tsp_with_path(graph):
    n = len(graph)
    dp = [[-1]*(1<<n) for _ in range(n)]
    parent = [[-1]*(1<<n) for _ in range(n)]

    def visit(city, mask):
        if mask == (1<<n) - 1:
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')
        next_city_choice = -1

        for next_city in range(n):
            if mask & (1<<next_city) == 0:
                new_cost = graph[city][next_city] + visit(next_city, mask | (1<<next_city))
                if new_cost < ans:
                    ans = new_cost
                    next_city_choice = next_city

        parent[city][mask] = next_city_choice
        dp[city][mask] = ans
        return ans

    cost = visit(0, 1)
    
    path = [0]
    city = 0
    mask = 1
    while mask != (1<<n) - 1:
        next_city = parent[city][mask]
        path.append(next_city)
        mask |= (1<<next_city)
        city = next_city
    path.append(0)
    
    return cost, path

graph_tsp = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

cost, route = tsp_with_path(graph_tsp)
print(f"Minimum cost: {cost}")
print(f"Route: {' -> '.join(map(str, route))}")

graph_6cities = [
    [0, 10, 12, 8, 15, 20],
    [10, 0, 5, 9, 12, 18],
    [12, 5, 0, 6, 8, 14],
    [8, 9, 6, 0, 7, 13],
    [15, 12, 8, 7, 0, 11],
    [20, 18, 14, 13, 11, 0]
]

cost_6, route_6 = tsp_with_path(graph_6cities)
print(f"\n6 cities - Minimum cost: {cost_6}")
print(f"Route: {' -> '.join(map(str, route_6))}")
```
