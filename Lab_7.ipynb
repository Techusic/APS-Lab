# Greedy Knapsack Problem (Fractional)
In this lab, we explore the Fractional Knapsack Problem, a fundamental optimization problem that demonstrates the Greedy Algorithmic Paradigm.
We will implement it using greedy selection (based on value-to-weight ratio) and analyze its time and space complexity.

# Problem Definition
Given a set of
n
n items, each with:
Value vi
Weight wi
A knapsack with maximum capacity **W**

Goal: maximize total value that can be put in the knapsack.
Constraint: You may take fractions of items.

Greedy Strategy:

*  Sort items by decreasing ratio vi/wi
*  Select items fully until knapsack is full.
*  Take fraction of the next item if capacity remains.

```python
#write code for the above problem

def fractional_knapsack(values, weights, capacity):
    if len(values) != len(weights):
        raise ValueError("values and weights must have the same length")
    if capacity < 0:
        raise ValueError("capacity must be non-negative")

    items = []
    for i, (v, w) in enumerate(zip(values, weights)):
        if w <= 0:
            raise ValueError(f"weight must be positive for item {i}")
        items.append((v / w, v, w, i))

    # Greedy choice: highest value-to-weight ratio first.
    items.sort(reverse=True, key=lambda x: x[0])

    total_value = 0.0
    selected = []
    remaining = float(capacity)

    for ratio, value, weight, idx in items:
        if remaining <= 0:
            break

        take_weight = min(weight, remaining)
        fraction = take_weight / weight
        gained_value = value * fraction

        selected.append({
            "item_index": idx,
            "fraction_taken": fraction,
            "weight_taken": take_weight,
            "value_gained": gained_value,
        })

        total_value += gained_value
        remaining -= take_weight

    return total_value, selected


# Example
values = [60, 100, 120]
weights = [10, 20, 30]
capacity = 50
best_value, selection = fractional_knapsack(values, weights, capacity)

print("Maximum value:", round(best_value, 2))
print("Selection details:")
for row in selection:
    print(row)

# Complexity
print("Time Complexity: O(n log n) due to sorting")
print("Space Complexity: O(n)")
```

In this lab, we implement Huffman Coding, a greedy algorithm used for lossless data compression.
It assigns variable-length binary codes to input symbols based on their frequencies, ensuring that more frequent symbols receive shorter codes and less frequent symbols receive longer codes.

This algorithm constructs a binary tree (Huffman Tree) where each leaf node represents a character, and the path from the root to a leaf gives the character’s binary code.

Use:
Huffman Tree Construction using a Min-Priority Queue (Min-Heap).
Code generation for each character.
Encoding and decoding using the generated Huffman codes.
Complexity analysis and comparison with fixed-length encoding.

# Pseudocode:
1. Create a leaf node for each character and insert all nodes into a
2. min-heap by frequency.
   *  While the heap has more than one node:
   * Extract two nodes with the lowest frequencies.
   * Create a new internal node with frequency = sum of the two.
   *  Set the two extracted nodes as left and right children.
   *  Insert the new node back into the min-heap.
3. The remaining node is the root of the Huffman Tree.

4. Traverse the tree:
   * Assign 0 for left edge and 1 for right edge.
   * The resulting paths from root to leaves give Huffman codes.

```python
# write code for above

import heapq
import math
from collections import Counter


class HuffmanNode:
    def __init__(self, char=None, freq=0):
        self.char = char
        self.freq = freq
        self.left = None
        self.right = None

    def __lt__(self, other):
        return self.freq < other.freq


def build_huffman_tree(text):
    if not text:
        return None

    frequency = Counter(text)
    heap = [HuffmanNode(ch, fr) for ch, fr in frequency.items()]
    heapq.heapify(heap)

    if len(heap) == 1:
        root = HuffmanNode(None, heap[0].freq)
        root.left = heap[0]
        return root

    while len(heap) > 1:
        left = heapq.heappop(heap)
        right = heapq.heappop(heap)

        merged = HuffmanNode(None, left.freq + right.freq)
        merged.left = left
        merged.right = right
        heapq.heappush(heap, merged)

    return heap[0]


def generate_codes(node, prefix="", code_map=None):
    if code_map is None:
        code_map = {}
    if node is None:
        return code_map

    if node.char is not None:
        code_map[node.char] = prefix if prefix else "0"
        return code_map

    generate_codes(node.left, prefix + "0", code_map)
    generate_codes(node.right, prefix + "1", code_map)
    return code_map


def huffman_encode(text, code_map):
    return "".join(code_map[ch] for ch in text)


def huffman_decode(bitstring, root):
    if root is None:
        return ""

    decoded = []
    node = root

    for bit in bitstring:
        node = node.left if bit == "0" else node.right
        if node.char is not None:
            decoded.append(node.char)
            node = root

    return "".join(decoded)


# Example
text = "huffman coding is a greedy algorithm"
root = build_huffman_tree(text)
codes = generate_codes(root)
encoded = huffman_encode(text, codes)
decoded = huffman_decode(encoded, root)

print("Original text:", text)
print("Huffman codes:", codes)
print("Encoded bits:", encoded)
print("Decoded text:", decoded)
print("Decoded matches original:", decoded == text)

# Compression comparison vs fixed-length coding
unique_chars = len(set(text))
fixed_bits_per_char = max(1, math.ceil(math.log2(unique_chars)))
fixed_length_bits = len(text) * fixed_bits_per_char
huffman_bits = len(encoded)

print("Fixed-length bits:", fixed_length_bits)
print("Huffman bits:", huffman_bits)
if fixed_length_bits > 0:
    savings = (fixed_length_bits - huffman_bits) / fixed_length_bits * 100
    print("Compression gain over fixed-length: {:.2f}%".format(savings))

# Complexity
print("Time Complexity: O(n log k), where n = text length and k = unique characters")
print("Space Complexity: O(k)")
```
